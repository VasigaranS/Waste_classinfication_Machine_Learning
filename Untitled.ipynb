```python
import tensorflow as tf
from tensorflow.keras import models,layers
```

```python
import matplotlib.pyplot as plt
```

```python
image_size=256
batch_size=32
```

```python
dataset=tf.keras.preprocessing.image_dataset_from_directory('DATASET/TRAIN',shuffle=True,image_size=(image_size,image_size),batch_size=batch_size)
test_dataset=tf.keras.preprocessing.image_dataset_from_directory('DATASET/TEST',shuffle=True,image_size=(image_size,image_size),batch_size=batch_size)

```

```python
class_names=dataset.class_names
class_names
```

```python
len(dataset)*0.89
```

```python
dataset=dataset.shuffle(100000,seed=20)
```

```python
train_ds=dataset.take(628)
```

```python
val_ds=dataset.skip(628)
```

```python
len(val_ds)
```

```python
len(test_dataset)
```

```python
#prefetch to read faster gpu trains and cpu fetches next batch
train_ds=train_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
val_ds=val_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
test_dataset=test_dataset.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
```

```python
#scaling to 0 and 1
#we alreday resized but while predicting in order to handle an image which is not 256,256, we have to add a layer
resize_rescale_layer=tf.keras.Sequential([layers.experimental.preprocessing.Resizing(image_size,image_size),
    layers.experimental.preprocessing.Rescaling(1.0/255)])
```

```python
#data augmentation
data_augmentation=tf.keras.Sequential([layers.experimental.preprocessing.RandomFlip("horizontal_and_vertical"),
                                      layers.experimental.preprocessing.RandomRotation(0.2)])
```

```python
model=models.Sequential([resize_rescale_layer,data_augmentation,
                         layers.Conv2D(32,(3,3),activation='relu',input_shape=(32,256,256,3)),
                         layers.MaxPooling2D((2,2)),
                         layers.Conv2D(64,(3,3),activation='relu'),
                         layers.MaxPooling2D((2,2)),
                         layers.Conv2D(64,(3,3),activation='relu'),
                         layers.MaxPooling2D((2,2)),
                         layers.Conv2D(64,(3,3),activation='relu'),
                         layers.MaxPooling2D((2,2)),
                         layers.Conv2D(64,(3,3),activation='relu'),
                         layers.MaxPooling2D((2,2)),
                         layers.Conv2D(64,(3,3),activation='relu'),
                         layers.MaxPooling2D((2,2)),
                         
                         layers.Flatten(),
                         layers.Dense(64,activation='relu'),
                         layers.Dense(2,activation='softmax'),
                         
                        
                        
                        ])
```

```python
model.build(input_shape=(32,256,256,3))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),metrics=['accuracy'])
```

```python
history=model.fit(train_ds,epochs=5,batch_size=32,verbose=1,validation_data=val_ds)
```
